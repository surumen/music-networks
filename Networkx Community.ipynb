```python
import matplotlib.pyplot as plt
%matplotlib inline
import networkx as nx

musicG = nx.read_edgelist("/Users/kuhukgoyal/Desktop/School/Networks_Final/final.csv", 
                     nodetype = str, delimiter = ',', data = (('sim_measure', float),))



```

```python
eList = list(musicG.edges(data='sim_measure', default=1)).copy()
print(len(eList))
for edge in eList:
    if(edge[2] < 0.50):
        musicG.remove_edge(*edge[:2])
print(len(musicG.edges()))
```

```python
music_giant = max(nx.connected_component_subgraphs(musicG), key=len)
```

```python
nx.draw_networkx(music_giant, with_labels=False)
plt.show()
```
